```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
q=0.5
b=0.9
w=2/3
```

```python
t=0
thetain=np.pi/3
win=1
y=np.array([thetain,win])
```

```python
def F(y,t):
    return np.array([y[1],-q*y[1]-np.sin(y[0])+b*np.cos(w*t)])   
```

```python
tsol=[t]
thetasol=[y[0]]
wsol=[y[1]]
dt=0.1
tfin=100
```

```python
while (t<tfin):
    k1=F(y,t)
    k2=F(y+dt*k1/2,t+dt/2)
    k3=F(y+dt*k2/2,t+dt/2)
    k4=F(y+dt*k3,t*dt)
    y=y+(k1+2*k2+2*k3+k4)*dt/6
    t=t+dt
    thetasol.append(y[0])
    wsol.append(y[1])
    tsol.append(t)
```

```python
plt.plot(thetasol,wsol,color='red')
plt.title('Grafico en el espacio(theta,w)')
plt.grid(True)
plt.xlabel('Theta')
plt.ylabel('w (velocidad angular)')
plt.savefig('P1-aguilarcarvajal')
```
