```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    f=1-np.sin(x)-np.cos(1+x**2)
    return f
```

```python
xp=np.linspace(-5, 5,1000)
plt.plot(xp,f(xp))
plt.grid(True)
```
